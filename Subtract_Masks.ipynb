```python
import sys
import os
import numpy as np
# from PIL import Image
import cv2
```

```python
relevant_path = os.getcwd()
Normal_mask_folder = "D:\\AB\\ASAP_Annotation_Mask\\OICR_Anno_Mask\\Normal"
Tumor_mask_folder = "D:\\AB\\ASAP_Annotation_Mask\\OICR_Anno_Mask\\Tumour"

Mask_extension = ['png']

Mask_List = [fn for fn in os.listdir(Tumor_mask_folder) if any(fn.endswith(ext) for ext in Mask_extension)]


Mask_output = os.path.join(relevant_path, "Final_Tumor_Masks")
if os.path.exists(Mask_output):
    print("Mask output directory already exists.")
else:
    os.mkdir(Mask_output)
```

```python
for M in Mask_List:
    print(M)
    Tumor = cv2.imread(os.path.join(Tumor_mask_folder, M), cv2.IMREAD_GRAYSCALE)
    Normal = cv2.imread(os.path.join(Normal_mask_folder, M), cv2.IMREAD_GRAYSCALE)
    
    _, Tumor = cv2.threshold(Tumor, 127, 255, cv2.THRESH_BINARY)
    _, Normal = cv2.threshold(Normal, 127, 255, cv2.THRESH_BINARY)
    
    Final_tumor = cv2.subtract(Tumor, Normal)
    
    cv2.imwrite(os.path.join(Mask_output, M), Final_tumor)
```
